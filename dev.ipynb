```python
import dataclasses, iso639
import pickle
from sklearn.metrics import classification_report
import pandas as pd

def get_iso_693_3(lang_id):
    if lang_id == 'jw':
        lang_id = 'jv'
    if lang_id == 'iw':
        return 'iw111'
    if lang_id == 'zho':
        return 'cmn'
    language = iso639.Language.match(lang_id)
    return dataclasses.asdict(language)['part3']

decode = {i: fleurs_langID.features["lang_id"].names[i].split('_')[0] for i in range(102)}
class_names = [get_iso_693_3(decode[s]) for s in range(102)]

y_true = pickle.load(open('whisper_results/whisper_true_labels.pkl', 'rb'))
y_pred = pickle.load(open('whisper_results/whisper_prediction.pkl', 'rb'))

# y_pred = []
# with open('voxlingua_results/voxlingua_pred.txt', 'r') as f:
#     for l in f:
#         y_pred.append(l.split()[-1])

print(len(y_true))
print(len(y_pred))

pd.set_option("display.max_rows", None)
print("whisper's prediction")
print(pd.Series(y_pred).value_counts())
print("testing set ground truth")
print(pd.Series(y_pred).value_counts())

y_pred = [get_iso_693_3(l) for l in y_pred]

print(classification_report(y_true, y_pred, labels=class_names, zero_division=0))
```
